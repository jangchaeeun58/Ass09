# 1. Graph Creation
# ============================================================

import networkx as nx
from collections import deque

G = nx.Graph()

nodes = [
    "상미규카츠",
    "파작",
    "몽탄",
    "그로어스",
    "아름당"
]

edges = [
    ("상미규카츠", "파작", 23),
    ("상미규카츠", "몽탄", 42),
    ("상미규카츠", "그로어스", 39),
    ("상미규카츠", "아름당", 43),

    ("파작", "몽탄", 22),
    ("파작", "그로어스", 31),
    ("파작", "아름당", 44),

    ("몽탄", "그로어스", 37),
    ("몽탄", "아름당", 25),

    ("그로어스", "아름당", 24)
]

G.add_nodes_from(nodes)

for u, v, w in edges:
    G.add_edge(u, v, weight=w)

print("그래프 생성 완료")

# ============================================================
# 2. DFS Algorithm
# ============================================================

visited_order = []
visited = set()

def dfs(node):

    visited.add(node)
    visited_order.append(node)

    for neighbor in G.neighbors(node):

        if neighbor not in visited:
            dfs(neighbor)

dfs("상미규카츠")

print("\nDFS 방문 순서")
print(visited_order)

# ============================================================
# 3. BFS Algorithm
# ============================================================

visit_order = []

def bfs(start):

    visited = set()
    queue = deque()

    visited.add(start)
    queue.append(start)

    while queue:

        current = queue.popleft()

        visit_order.append(current)

        for neighbor in G.neighbors(current):

            if neighbor not in visited:

                visited.add(neighbor)
                queue.append(neighbor)

bfs("상미규카츠")

print("\nBFS 방문 순서")
print(visit_order)

# ============================================================
# 4. Prim Algorithm
# ============================================================

mst_edges = []
visited = set()

start_node = "상미규카츠"
visited.add(start_node)

while len(visited) < len(G.nodes()):

    min_edge = None
    min_weight = float('inf')

    for u in visited:

        for v in G.neighbors(u):

            if v not in visited:

                weight = G[u][v]['weight']

                if weight < min_weight:

                    min_weight = weight
                    min_edge = (u, v, weight)

    if min_edge:

        u, v, w = min_edge

        visited.add(v)
        mst_edges.append((u, v, w))

print("\nPrim MST")
print(mst_edges)

# ============================================================
# 5. Kruskal Algorithm
# ============================================================

parent = {}

for node in G.nodes():
    parent[node] = node

def find(node):

    if parent[node] != node:
        parent[node] = find(parent[node])

    return parent[node]

def union(a, b):

    rootA = find(a)
    rootB = find(b)

    if rootA != rootB:
        parent[rootB] = rootA

sorted_edges = sorted(edges, key=lambda x: x[2])

mst_edges = []

for u, v, w in sorted_edges:

    if find(u) != find(v):

        union(u, v)
        mst_edges.append((u, v, w))

print("\nKruskal MST")
print(mst_edges)
